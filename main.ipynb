## 1. Load the results of experiments

```python
%matplotlib inline

import glob

import matplotlib.pyplot as plt
import pandas as pd
```

```python
array_blocking_queue_results_file = glob.glob('input/ArrayBlockingQueue*.csv')[0]
abq_df = pd.read_csv(array_blocking_queue_results_file)
linked_blocking_queue_results_file = glob.glob('input/LinkedBlockingQueue_high_bound*.csv')[0]
lbq_df = pd.read_csv(linked_blocking_queue_results_file)
linked_blocking_queue_low_bound_results_file = glob.glob('input/LinkedBlockingQueue_low_bound*.csv')[0]
lbq_low_bound_df = pd.read_csv(linked_blocking_queue_low_bound_results_file)
linked_concurrent_queue_results_file = glob.glob('input/LinkedConcurrentQueue*.csv')[0]
lcq_df = pd.read_csv(linked_concurrent_queue_results_file)

x_axis_data = abq_df['Number of messages']
```

## 2. Visualize results

```python
fig, ax = plt.subplots()
ax.set_title('Comparison of TokenRing implementations (median circle latency)', y=1.1)
ax.set_xlabel('Number of messages')
ax.set_ylabel('Latency')
ax.plot(x_axis_data, abq_df["Median latency"], label='ArrayBlockingQueue')
ax.plot(x_axis_data, lbq_df["Median latency"], label='LinkedBlockingQueue')
ax.plot(x_axis_data, lcq_df["Median latency"], label='LinkedConcurrentQueue')
ax.legend()
plt.show()

fig.savefig('output/latency.png', bbox_inches='tight', pad_inches=2)
```

```python
fig, ax = plt.subplots()
ax.set_title('Comparison of TokenRing implementations (median throughput per second)', y=1.1)
ax.set_xlabel('Number of messages')
ax.set_ylabel('Throughput')
ax.plot(x_axis_data, abq_df["Median throughput"], label='ArrayBlockingQueue')
ax.plot(x_axis_data, lbq_df["Median throughput"], label='LinkedBlockingQueue')
ax.plot(x_axis_data, lcq_df["Median throughput"], label='LinkedConcurrentQueue')
ax.legend()
plt.show()

fig.savefig('output/throughput.png', bbox_inches='tight', pad_inches=2)
```

We can also analyze results separately for each implementation.

```python
fig, ax = plt.subplots()
ax.set_title('TokenRing implementation using LinkedBLockingQueue with different capacity (median circle latency)', y=1.1)
ax.set_xlabel('Number of messages')
ax.set_ylabel('Latency')
ax.plot(x_axis_data, lbq_df["Median latency"], label='N = 4 and capacity = 100')
ax.plot(x_axis_data, lbq_low_bound_df["Median latency"], label='N = 4 and capacity = 35')
ax.legend()
plt.show()

fig.savefig('output/linked_queue_latency.png', bbox_inches='tight', pad_inches=2)
```

```python
fig, ax = plt.subplots()
ax.set_title('TokenRing implementation using LinkedBLockingQueue with different capacity (median throughput per second)', y=1.1)
ax.set_xlabel('Number of messages')
ax.set_ylabel('Throughput')
ax.plot(x_axis_data, lbq_df["Median throughput"], label='N = 4 and capacity = 100')
ax.plot(x_axis_data, lbq_low_bound_df["Median throughput"], label='N = 4 and capacity = 35')
ax.legend()
plt.show()

fig.savefig('output/linked_queue_throughput.png', bbox_inches='tight', pad_inches=2)
```
